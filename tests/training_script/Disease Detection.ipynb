```python
numeric_columns = ['age', 'bp', 'sg', 'al', 'bgr','bu','sc','sod','pot','hemo','pcv', 'wbcc', 'rbcc']
```

```python
cat_cols = []
for c in list(df.columns):
    if c not in numeric_columns:
        cat_cols.append(c)
```

```python
cat_cols = cat_cols[1:-1]
```

```python
cat_cols
```

```python
for n in numeric_columns:
    df[n] = df[n].astype(str).astype(float)
```

```python
le = LabelEncoder()
for c in cat_cols:
    df[c] = le.fit_transform(df[c])
```

```python
df
```

```python
df.info()
```

```python
df['class'].unique()
```

```python
class_dict = {'ckd':1, 'notckd':0}
```

```python
df['class'] = df['class'].map(class_dict)
```

```python
df
```

```python
X = df.drop(['class'], axis=1)
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

### Random Forest Model

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
```

### Model Evaluation

```python
accuracy_score(y_test, preds)
```

```python
print(classification_report(y_test, preds))
```

```python
import pickle as pkl

filename = '../model/kidney_disease.pkl'
pkl.dump(model, open(filename, 'wb'))

pd.concat([X_test,y_test],axis=1).to_csv(f"../data/split/test_data.csv", index=None,)
X_test.to_csv(f"../data/split/test_data_no_class.csv", index=False)
X_test.to_csv(f"../data/split/test_data_no_class.txt", index=False, line_terminator="\\n")
pd.concat([X_test,y_test],axis=1).to_json(f"../data/split/test_data_json.json", orient="table", index=False)
X_test.to_json(f"../data/split/test_data_json_no_class.json", orient="table", index=False)
```
